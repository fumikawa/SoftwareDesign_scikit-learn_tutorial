```python
# 必要なライブラリのインポート
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# データセットの準備
X, y = make_blobs(n_samples=1500, n_features=2, centers=2, random_state=2)

# 特徴量（横軸）のスケールを拡大
X[:, 0] = X[:, 0] * 10

# データの分布傾向の可視化
plt.scatter(X[:, 0], X[:, 1])
plt.title("Blobs Data ")
plt.show()
```

```python
# 必要なライブラリのインポート
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

# データの正規化
X_norm = MinMaxScaler().fit_transform(X)

# k-meansの適用（クラスタ数は2）
kmeans = KMeans(n_clusters=2, random_state=0)
y_pred = kmeans.fit_predict(X_norm)

# クラスタリング結果の可視化
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=~y_pred)
plt.title("(a) k-means cluster")
plt.subplot(122)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=y)
plt.title("(b) true cluster")
plt.show()
```

```python
# k-meansの適用（クラスタ数は2）
kmeans = KMeans(n_clusters=2, random_state=0)
y_pred = kmeans.fit_predict(X)

# クラスタリング結果の可視化
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("(a) k-means cluster")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("(b) true cluster")
plt.show()
```
